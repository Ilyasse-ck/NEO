```python
#Creation d'un graph et de la creation de la matriz d'incidence:
Graph={}

for index , row in Donnes.iterrows():
    if row[0] in Graph:
        Graph[int(row[0])].append([int(row[1]) ,row[2]])
    else:
        Graph[int(row[0])]=[[int(row[1]),row[2]]]
        
#Creation d'une Matriz d'incidense
M_inc=[]
for i in range(len(Graph)):
    M_inc.append([])
    for j in range(len(Graph)):
        M_inc[i].append(0)

for i in Graph:
    for arc in Graph[i]:
        M_inc[i-1][arc[0]-1]=1
        M_inc[arc[0]-1][i-1]=-1
```

### Algorithme de Dijsktra

```python
def Dijsktra(Graph,debut,fin):
    D=[1000]   #Pour conter les index=valeur du sommet
    M=[0]
    for i in Graph:
        if i==debut:
            D.append(0)
            M.append(0)
        else:
            D.append(1000)
            M.append(-1)
    Q=[debut]
    while len(Q)!=0:
        # Je vais chercher le sommet qu'il y a la distance minimale sur la list Q
        Dmin=[D[i] for i in Q]
        dmin=min(Dmin)
        index=Dmin.index(dmin)
        i=Q[index]
        Q.remove(i)
        # Pour charque arc qui sort du sommet selectione on va calculer la distance 
        for arc in Graph[i]:
            j=arc[0]
            if D[j]>(D[i]+arc[1]):
                D[j]=(D[i]+arc[1])
                M[j]=i
                if not j in Q:
                    Q.append(j)
                
    return [D[fin],M]
        
    
```

La on va essayer d'améliorer l'algorithme de Dijsktra en faisant une recherche concurente:
->On va faire un plus court chemin de s à t et au même temp de t à s, et a la fin on va melanger les deux codes

```python
def Dijsktra_Ameliore(Graph,M_inc,debut,fin):
    D=[1000]   #Pour conter les index=valeur du sommet
    M=[-1]
    N=[-1]
    Dc=[1000]
    for i in Graph:
        if i==debut:
            D.append(0)
            M.append(0)
            Dc.append(1000)
            N.append(-1)
            
        elif i==fin:
            D.append(1000)
            M.append(-1)
            Dc.append(0)
            N.append(0)
            
        else:
            D.append(1000)
            M.append(-1)
            Dc.append(1000)
            N.append(-1)
    Q=[debut]
    Qc=[fin]
    DMAX=1000
    while len(Q)!=0 and len(Qc)!=0:
        # Je vais chercher le sommet qu'il y a la distance minimale sur la list Q
        Dmin=[D[i] for i in Q]
        dmin=min(Dmin)
        index=Dmin.index(dmin)
        i=Q[index]
        Q.remove(i)
        # Pour charque arc qui sort du sommet selectione on va calculer la distance 
        for arc in Graph[i]:
            j=arc[0]
            if D[j]>(D[i]+arc[1]):
                D[j]=(D[i]+arc[1])
                M[j]=i
                if not j in Q and j!=fin and D[j]<=DMAX:# Si le cout est dejá assez grand on arrete
                    Q.append(j)
                
        # La on va faire la même analyse mais avec un chemin contraire et pour savoir quelles sont les
        #sommets qui pouvent acceder le sommets actuel.
        # Je vais chercher le sommet qu'il y a la distance minimale sur la list Q
        Dmin=[Dc[i] for i in Qc]
        dmin=min(Dmin)
        index=Dmin.index(dmin)
        A=Qc[index]
        Qc.remove(A)
        # Pour charque arc qui arrive au sommet selectione on va calculer la distance 
        for j in range(len(M_inc)):
            if M_inc[A-1][j]==-1: #Il'y a un arc qui sort de (j+1) et arrive a i
                p=0
                for arc in Graph[j+1]:
                    if arc[0]==A:
                        position=p
                    p+1
                if Dc[j+1]>(Dc[A]+Graph[j+1][position][1]):
                    Dc[j+1]=(Dc[A]+Graph[j+1][position][1])
                    N[j+1]=A
                    
                if not (j+1) in Qc and (j+1) !=debut and Dc[j+1]<=DMAX: # Si le cout est dejá assez grand on arrete
                    Qc.append(j+1)
        # Là, il y faut unir les chemins et voir si on aura déjà une chemin, pas le optimale entre les deux,
        for i in range(len(M)):
            if M[i]!=-1 and N[i]!=-1: # Il y a un chemin du debut au sommet et un autre du sommet à la fin
                Chemin=D[i]+Dc[i]
                if DMAX>Chemin:
                    DMAX=Chemin
           
    #Pour donner le chemin optimale:
    Chemin=[fin]
    
    while Chemin[0]!=debut:
        dernier=Chemin[0]
        Chemin=[M[dernier]]+Chemin
    return [D[fin],Chemin]
        
```

### Algorithme A*

Comme fonction de coût, j'ai choisi d'utiliser une moyenne quadratique de toutes les destinations possibles du sommet qui peut être choisi.

```python
def h(i,Graph):
    Cout=0
    for arc in Graph[i]:
        Cout = Cout+arc[1]**2
    
    return np.sqrt(Cout/len(Graph[i]))
```

```python
def A_etoile(Graph,debut,fin):
    D=[1000]   #Pour conter les index=valeur du sommet
    M=[0]
    for i in Graph:
        if i==debut:
            D.append(0)
            M.append(0)
        else:
            D.append(1000)
            M.append(-1)
    Q=[debut]
    while len(Q)!=0:
        # Je vais chercher le sommet qu'il y a la distance minimale sur la list Q
        Dmin=[(D[i]+h(i,Graph)) for i in Q]
        dmin=min(Dmin)
        index=Dmin.index(dmin)
        i=Q[index]
        Q.remove(i)
        # Pour charque arc qui sort du sommet selectione on va calculer la distance 
        for arc in Graph[i]:
            j=arc[0]
            if D[j]>(D[i]+arc[1]):
                D[j]=(D[i]+arc[1])
                M[j]=i
                if not j in Q:
                    Q.append(j)
                
    #Pour donner le chemin optimale:
    Chemin=[fin]
    
    while Chemin[0]!=debut:
        dernier=Chemin[0]
        Chemin=[M[dernier]]+Chemin
    return [D[fin],Chemin]
```

Utilisons maintenant la hiérarchie dans le calcul, comme s'il s'agissait d'un arbre. Nous donnerons donc toujours la priorité aux sommets qui se trouvent autour du sommet actuel. S'il n'y a pas de sommets à calculer, nous passons à une autre branche de l'arbre, un sommet de la liste de questions précédente.

```python
def A_etoile_heuristique(Graph,debut,fin):
    D=[1000]   #Pour conter les index=valeur du sommet
    M=[0]
    for i in Graph:
        if i==debut:
            D.append(0)
            M.append(0)
        else:
            D.append(1000)
            M.append(-1)
    Q=[debut]
    Q_voysinage=[]
    while len(Q)!=0:
        # Je vais chercher le sommet qu'il y a la distance minimale sur la list Q
        # On va donner une priorité pour les sommets qui sont sur la voysinage du sommet actuele.
        if len(Q_voysinage)!=0:
            Dmin=[(D[i]+h(i,Graph)) for i in Q_voysinage]
            dmin=min(Dmin)
            index=Dmin.index(dmin)
            i=Q_voysinage[index]
            Q.remove(i)
        # Si il'y a pas de voysinage on va arreter lá
        else:
            Dmin=[(D[i]+h(i,Graph)) for i in Q]
            dmin=min(Dmin)
            index=Dmin.index(dmin)
            i=Q[index]
            Q.remove(i)
        # Pour charque arc qui sort du sommet selectione on va calculer la distance 
        Q_voysinage=[]
        for arc in Graph[i]:
            j=arc[0]
            if D[j]>(D[i]+arc[1]):
                D[j]=(D[i]+arc[1])
                M[j]=i
                Q_voysinage.append(j)
                if not j in Q:
                    Q.append(j)
                
    Chemin=[fin]
    
    while Chemin[0]!=debut:
        dernier=Chemin[0]
        Chemin=[M[dernier]]+Chemin
    return [D[fin],Chemin]
                
```
